```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv('/content/movies (1) (1).csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
movie=movies[["name","rating","genre","score","director","star","country"]]
```

```python
movie=movie.dropna()
```

```python
movie.shape
```

```python
movie.isnull().sum()
```

```python
movie_feature=movie["name"]+' '+movie["rating"]+' '+movie["genre"]+' '+movie["director"]+' '+movie["star"]+' '+movie["country"]
movie_feature
```

```python
vectorizer=TfidfVectorizer()
```

```python
movie_vectorizer=vectorizer.fit_transform(movie_feature)
```

```python
similarity=cosine_similarity(movie_vectorizer)
```

```python
## creating a user defind function for movie recomandation.
movie_name = input(' Enter your favourite movie name : ')


# creating a list with all the movie names given in the data.
list_of_titles = movies['name'].tolist()


# finding the close match for the movie name given by the user with the help of difflib function.

finds_close_match = difflib.get_close_matches(movie_name, list_of_titles)


close_match = finds_close_match[0]

# find the index of the movie with title of movie.

index_of_movie = movies[movies.name == close_match]['index'].values[0]


similarity_score = list(enumerate(similarity[index_of_movie]))

# sorting the movies based on their similarity score.

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies[movies.index==index]['name'].values[0]
  if (i<10):
    print(i, '.',title_from_index)
    i+=1
```
